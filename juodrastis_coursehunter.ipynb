```python
#tuple
skaiciai = (10, 20)
sarasas = list(skaiciai) #kad veiktu append reikia sukonvertuoti i sarasa t.y prideti "list"
sarasas.append(30)
print(sarasas) #[10, 20, 30] sarasas [] 

#sukonvertuoti atgal i kortezus reikia sarasa paversti i tuple:
sarasas_tuple = tuple(sarasas)
print(sarasas_tuple)  # (10, 20, 30) () grizo i tuple

#count
skaiciai = (10, 20, 30, 20)
print(skaiciai.count(10)) #suskaiciuoja kiek kartu parirode skaicius 10. t.y 1 karta
print(skaiciai.count(20)) #suskaiciuoja kiek kartu parirode skaicius 20. t.y 2 kartus

#index
skaiciai = (10, 20, 30, 20)
print(skaiciai.index(10)) #parodo eiles numeri indexa sarase. skaicius 10, yra 0 sarase
print(skaiciai.index(20)) #parodo eiles numeri indexa sarase. skaicius 20, yra 1 sarase, tie patys skaiciai
# einantys sarase neaktulus

#uzduotis: istrinti my_nums = (10, 5, 100, 0) skaiciu pvz 10

my_nums = (10, 5, 100, 0) #tuple

# --del my_nums[0]--- # bandom istrinti skaiciu 10: TypeError: 'tuple' object doesn't support item deletion
#konvertuojam tuple i sarasa

my_nums_list = list(my_nums) #konvertuojam i list
del my_nums_list[0] #istrinam
print(my_nums_list) # [5, 100, 0], bet atsakymas sarasas []
my_nums_list = tuple(my_nums_list) # konvertuojam atgal i tuple
print(my_nums_list) # (5, 100, 0)

#kovertuojam tuple i sarasa arba zodyna:
mano_tuple = tuple('abcd') #gali buti tik 1 argumentas ir tik kabutese! pvz: '10' 'b'
print(mano_tuple) #('a', 'b', 'c', 'd') i sarasa
mano_tuple2 = tuple({'modelis': 'Audi', 'variklis': 'dyzelis'})
print(mano_tuple2) # grazino spynas ('modelis', 'variklis') zadziant su zodynu

#uzduotis sukurti du sarasaus ir sujunti du tuples juos sudedant
saras1 = tuple('10')
saras2 = tuple('b')
naujas_sar = saras1 + saras2
print(naujas_sar) #('1', '0', 'b')

```

```python
# ZIP sarasu [] sujungimas
vaisiai = ['obuoliai', 'kriauses']
kaina = [2, 8]
vaisiai_kaina_zip = zip(vaisiai, kaina)
print(vaisiai_kaina_zip) # gaunam <zip object at 0x000001D7E6DDA600>
print(list(vaisiai_kaina_zip)) #tada reikia konvertuoti i lista:
# [('obuoliai', 2), ('kriauses', 8)] 

#arba galima perdaryt i zodyna {}
vaisiai1 = ['obuoliai', 'kriauses']
kaina1 = (2, 8) #galima naudoti ir kitus variantus () [] {} kaip sarasa ir pan
vaisiai_kaina_zip1 = zip(vaisiai1, kaina1)
print(dict(vaisiai_kaina_zip1)) # gauname kai zodyno varianta: {'obuoliai': 2, 'kriauses': 8}
```

```python
#CIKLAI!!!! DICT - ZODYNAI {}, LIST - SARASAI [], TUPLE() NEGALI BUTI MODIFIKUOTAS (NEI TRINT NIEKO), SET - AIBES {}, RANGE, STR
# DU TIPAI for... in.. while

#sarasas []
mano_sarasas_list = [True, 10, 'abs'] #gali buti sarase bet kokie tipai
for elementai in mano_sarasas_list:
    print(elementai) # atspausdina viskas kas yra mano sarase

#tuple ()
mano_sarasas_tuple = (True, 10, 'abs') #gali buti sarase bet kokie tipai bet tvarka sarase turi itakos
for elementai in mano_sarasas_tuple:
    print(elementai)

#zodynas {}
zmoniu_info = {'Andrius': 20, 'Rasa': 40} #vardas, amzius
for isvestis in zmoniu_info: #
    print(isvestis, zmoniu_info[isvestis]) # Andrius 20 , Rasa 40
#antras variantas
for vardas, metai in zmoniu_info.items(): #tai pat isves rezultata jeigu nurodysim items()
    print(vardas, metai)  # # Andrius 20 , Rasa 40

#range

for skaiciai in range(5):
    print(skaiciai +1)
```
